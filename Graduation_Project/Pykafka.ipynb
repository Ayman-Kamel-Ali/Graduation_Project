```python
import json
import random
import time
from kafka import KafkaProducer

```

### Kafka configuration


```python
KAFKA_BROKER = 'localhost:9092'  
TOPIC_NAME = 'heart'
```

### Simulating Sensor Data


```python
def generate_sensor_data():
    
    
    return {

        "patient_id": 909090909090909090,
        "age": 40,
        "gender": 1, 
        "chestpaintype": 2,
        "restingBP": random.randint(90, 120), 
        "cholestrol": random.randint(150, 200),
        "fastingbloodsugar": 0,
        "restingECG": 1,
        "maxheartrate": random.randint(70, 90),
        "exerciseangia": 1,
        "oldpeak": round(random.uniform(0.0, 6.0), 1),
        "slope": 2

    }
```

### Initialize Kafka producer


```python
producer = KafkaProducer(
    bootstrap_servers=KAFKA_BROKER,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')  
)
```

### Sending data to Kafka


```python
try:
    while True:
        record = generate_sensor_data()  
        producer.send(TOPIC_NAME, value=record)  
        print(f"Sent record to Kafka: {record}")  
        time.sleep(3)  

except KeyboardInterrupt:
    print("Data generation stopped.")

finally:
    
    producer.close()
```
